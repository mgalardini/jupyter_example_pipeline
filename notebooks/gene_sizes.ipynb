Q: Is there a difference in gene size between strands?
-------------------------------------------------------

```python
fig_prefix = 'gene_sizes_'
```

```python
# plotting imports
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')

plt.rc('font', size=15)
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rc('axes', labelsize=16, titlesize=16)
plt.rc('legend', fontsize=15)
```

```python
# other imports
import pandas as pd
```

```python
t = pd.read_table('../gene_table.tsv',
                  header=None)
t.columns = ['replicon', 'gene', 'strand', 'size']
```

```python
plt.figure(figsize=(4, 7))

sns.stripplot(data=t,
            x='strand',
            y='size',
            jitter=0.35,
            alpha=0.35,
            palette=[sns.xkcd_rgb['pale red'],
                     sns.xkcd_rgb['sky blue'],])

plt.ylim(0, 8000)
plt.ylabel('gene size')

plt.xticks([0, 1],
           ['negative', 'positive'])

sns.despine(bottom=True)
plt.tight_layout()
plt.savefig('%s_1.pdf'%fig_prefix, dpi=300)
```

```python
plt.figure(figsize=(7, 3))

sns.kdeplot(t[t.strand == -1]['size'],
            color=sns.xkcd_rgb['pale red'],
            label='negative',
            lw=3)
sns.kdeplot(t[t.strand == 1]['size'],
            color=sns.xkcd_rgb['sky blue'],
            label='positive',
            lw=3)

plt.xlabel('gene size')
plt.ylabel('frequency')

sns.despine()
plt.savefig('%s_2.pdf'%fig_prefix, dpi=300)
```
